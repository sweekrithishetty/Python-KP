# Sets

```python
#Defining an empty set: #Not duplicate elements,Doesn't support indexing
a = set()
print(a)
```

```python
var = {'a','b','c'}
var
```

```python
#Inbuit functions
var.add('aa')
var
```

```python
set1 = {'a','b','c','d'}
set2 = {'a','b','c','d','e'}
set2.difference(set1)
set2
```

```python
set2.difference_update(set1) #It will update the set
print(set2)
```

```python
#Intersections
set1 = {'a','b','c','d'}
set2 = {'a','b','c','d','e'}
set1.intersection(set2)
```

```python
set1.intersection_update(set2)
set1
```

```python
#Dictionaries >> It has keys and values
dict = {}
type(dict)
```

```python
dic = {1,2,3,4,5}
type(dic)
```

```python
#Create a dictionary
my_dicr = {'1':'key','2':'values'}
my_dicr['1'] #Indexing is done via assigning key value.
```

```python
#We can loop through out the dictionaries keys
for x in my_dicr:
    print(x)
```

```python
#We can loop through out the dictionaries values
for x in my_dicr.values():
    print(x)
```

```python
#We can check both keys and values
for x in my_dicr.items():
    print(x)
```

```python
#Adding items in dictionary
my_dicr['3'] = 'both'
```

```python
my_dicr
```

```python
#Replacing the items
my_dicr['1'] = 'keyyys'
```

```python
my_dicr
```
